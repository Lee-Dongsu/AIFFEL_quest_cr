```python
import sys, math  # 프로그램 종료를 위한 sys 모듈과, math 모듈을 가져옵니다.

# 네가지의 연산 함수 정의
def add(a,b):
  return a + b

def subtract(a,b):
  return a - b

def multiply(a,b):
  return a * b

def divide(a,b):
  if b == 0:
    raise ZeroDivisionError('숫자 0으로 나눌 수 없습니다.')   # 0으로 나눌 경우의 에러 발생 -> 예외 처리 ZeroDivisionError 적용
  return f'몫 : {a // b}, 나머지 : {a % b}, 소수점 : {a / b}'

def power(a,b):
  return math.pow(a,b)    # math 모듈을 통해 제곱 연산을 반환 받습니다.

#계산을 반복적으로 수행 할 수 있도록 적용
while True:
  try:

    # 사용자로부터 숫자와 연산자 입력 받기
    a = int(input('첫 번째 숫자를 입력하세요 : '))
    b = int(input('두 번째 숫자를 입력하세요 : '))

    operation = input('연산자를 입력하세요 (+, -, *, /, **) : ')

    # 입력한 연산자에 알맞은 결과 출력
    if operation == '+':
      print(f'계산 결과 : {add(a,b)}')
    elif operation == '-':
      print(f'계산 결과 : {subtract(a,b)}')
    elif operation == '*':
      print(f'계산 결과 : {multiply(a,b)}')
    elif operation == '/':
      print(f'계산 결과 : {divide(a,b)}')
    elif operation == '**':
      print(f'계산 결과 : {power(a,b)}')
    else:
      print('잘못된 연산자를 입력했습니다. 다시 입력해주세요.')

  # 숫자가 아닌 값을 입력했을때의 예외 처리
  except ValueError:
    print('Error : 잘못된 입력입니다. 정수를 입력해주세요.')

  # 숫자를 0으로 나누었을때의 예외 처리
  except ZeroDivisionError as e:
    print(f'Error : {e} 다시 입력해주세요.')

  # 계산 종료 이후 계산을 계속 진행할지 입력 받습니다.
  while True:
    continue_cal = input('계산을 계속 진행 하시겠습니까? (y/n) : ')
    if continue_cal == 'y':
      break
      # 'y'를 입력시 루프를 빠져나와 다시 계산합니다.

    elif continue_cal == 'n':
      print('계산을 종료합니다.')
      sys.exit()
      # 'n'을 입력시 프로그램을 종료합니다.

    else:
      print('잘못된 입력입니다. 다시 입력해주세요.')
```
