```python
import random

# 클래스 캐릭터
class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    # 현재 체력이 0 이상인지를 나타내주는 메서드
    def is_alive(self):
        return self.hp > 0

    # 공격을 받았을 때의 메서드
    def take_damage(self, damage):
        actual_damage = max(0, damage - self.defense)
        self.hp = max(0, self.hp - actual_damage)
        print(f'{self.name} 이(가) {actual_damage}의 데미지를 받았습니다. 남은 체력 : {self.hp}')

    # 데미지를 입히는 메서드
    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        print(f'{self.name} 이(가) {target.name}에게 {damage}의 데미지를 입혔습니다.')
        target.take_damage(damage)

# 클래스 플레이어
class Player(Character):
    def __init__(self, name):
        # 속성 초기화
        super().__init__(name, 1, 100, 25, 5)
        self.exp = 0

    # 경험치를 획득하는 메서드
    def gain_experience(self, amount):
        self.exp += amount
        print(f'{self.name} 이(가) {amount}의 경험치를 얻었습니다. 현재 경험치 : {self.exp}')
        self.check_level_up()

    # 경험치 확인 및 레벨업 메서드
    def check_level_up(self):
        while self.exp >= 50:
            self.exp -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5
            print(f'{self.name} 이(가) 레벨업! 현재 레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense}')

# 클래스 몬스터
class Monster(Character):
    def __init__(self, name, level):
        # 몬스터 레벨에 비례한 스탯 초기화
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, hp, attack, defense)

# 전투 함수
def battle(player, monster):
    print('전투 시작!')
    # 플레이어, 몬스터가 살아 있을시 플레이어가 몬스터를 공격
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        # 이후 몬스터가 살아 있을시 플레이어를 공격
        if monster.is_alive():
            monster.attack_target(player)

    # 전투 후 플레이어가 살았다면 경험치 획득과 함께 전투 종료, 죽었다면 패배
    if player.is_alive():
        player.gain_experience(monster.level * 20)
        print('전투 승리!')
    else:
        print('전투 패배..')

# 메인 함수
def main():
    # 몬스터 딕셔너리 정의
    monster_dict = {'달팽이' : 1, '슬라임' : 2, '돼지' : 3, '머쉬맘' : 10}

    # 사용자 이름 설정
    player_name = input('사용할 이름 정해주세요 : ')
    player = Player(player_name)

    # 몬스터의 이름과 레벨 설정
    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)

        # 전투
        battle(player, monster)

        if not player.is_alive():
            print('게임오버')
            return

main()
```
