```python
import random

class Account:

    account_count = 0

    def __init__(self, name, balance):
        # 입금 횟수, 입출금 내역
        self.deposit_count = 0
        self.deposit_log = []
        self.withdraw_log = []
        self.balance_log = []

        # 계좌 정보 입력
        self.name = name
        self.balance = balance
        self.bank = 'SC은행'

        # 랜덤한 계좌번호 생성
        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)

        num1 = str(num1).zfill(3)
        num2 = str(num2).zfill(2)
        num3 = str(num3).zfill(6)
        self.account_number = num1 + '-' + num2 + '-' + num3

        Account.account_count += 1

    # Account 클래스로부터 생성된 계좌의 개수 출력
    @classmethod
    def get_account_num(cls):
        print('생성된 계좌의 개수 : ', cls.account_count)

    # 입금 메서드
    def deposit(self, amount):
        if amount >= 1:
            self.deposit_log.append(amount)
            self.balance += amount
            self.balance_log.append(self.balance)  # 입금 후 잔액 기록
            print(f'입금 금액: {amount:,}원, 잔액: {self.balance:,}원')

            # 입금 횟수 5회마다 이자 지급
            self.deposit_count += 1
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += interest
                self.deposit_log.append(interest)  # 이자 입금 기록
                self.balance_log.append(self.balance)  # 이자 입금 후 잔액 기록
                print(f'이자 입금 금액: {interest:.0f}원, 잔액: {self.balance:,}원')

    # 출금 메서드
    def withdraw(self, amount):
        if self.balance > amount:
            self.withdraw_log.append(amount)
            self.balance -= amount
            self.balance_log.append(self.balance)  # 출금 후 잔액 기록
            print(f'출금 금액: {amount:,}원, 잔액: {self.balance:,}원')

    # 정보 출력 메서드
    def display_info(self):
        print('은행이름 : ', self.bank)
        print('예금주 : ', self.name)
        print('계좌번호 : ', self.account_number)
        print('잔고 : ', f'{self.balance:,}')

    # 입금 기록 메서드
    def deposit_history(self):
        for i, amount in enumerate(self.deposit_log):
            print(f'{i+1}. 입금 금액: {amount:,}원, 잔액: {self.balance_log[i]:,}원')

    # 출금 기록 메서드
    def withdraw_history(self):
        deposit_length = len(self.deposit_log)
        for i, amount in enumerate(self.withdraw_log):
            print(f'{i+1}. 출금 금액: {amount:,}원, 잔액: {self.balance_log[deposit_length + i]:,}원')


# 객체 생성 및 저장
data = []
KIM = Account('김갑룡', 30000000)
LEE = Account('이나영', 2000000)
PARK = Account('박지성', 10000)

data.append(KIM)
data.append(LEE)
data.append(PARK)

Account.get_account_num()

# 객체 순회를 통해 잔고 100만원 이상의 고객만 출력
for c in data:
    if c.balance >= 1000000:
        c.display_info()



DONGSU = Account('이동수', 1000000)
print()
DONGSU.deposit(1000000)
DONGSU.deposit(1000000)
DONGSU.deposit(1000000)
DONGSU.deposit(1000000)
DONGSU.deposit(1000000)  # 이자 발생
DONGSU.withdraw(500000)
DONGSU.withdraw(500000)
print()
DONGSU.display_info()
print()
DONGSU.deposit_history()
print()
DONGSU.withdraw_history()
```
