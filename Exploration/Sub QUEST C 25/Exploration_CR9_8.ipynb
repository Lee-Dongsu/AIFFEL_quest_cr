### STEP 3. 생성자 모델 구현하기

```python
def make_generator_model():
    # 생성자 모델을 만듭니다.
    
    # Start
    model = tf.keras.Sequential()  # 순차적으로 층을 쌓는 Keras Sequential 모델을 생성합니다.

    # First: Dense layer
    # units은 8x8x256, 편향은 사용하지 않음, 입력 값의 크기는 100차원 벡터입니다.
    model.add(layers.Dense(8 * 8 * 256, use_bias=False, input_shape=(100,)))
    # 배치 정규화 사용하여 데이터 분포를 정규화합니다.
    model.add(layers.BatchNormalization())
    # LeakyReLU 활성화 함수 사용으로 음수 입력에 대해 작은 기울기를 허용합니다.
    model.add(layers.LeakyReLU())

    # Second: Reshape layer
    # 8x8 크기의 특징 맵 256개로 변환합니다. 
    model.add(layers.Reshape((8, 8, 256)))

    # Third: Conv2DTranspose layer
    # 필터의 수는 128개, 커널 크기는 5, 스트라이드는 1, 패딩을 사용하여 출력 크기를 동일하게 유지합니다.
    model.add(layers.Conv2DTranspose(128, kernel_size=5, strides=1, padding='same', use_bias=False))
    # 배치 정규화를 사용하여 학습을 안정화합니다.
    model.add(layers.BatchNormalization())
    # LeakyReLU 활성화 함수로 비선형성을 추가합니다.
    model.add(layers.LeakyReLU())

    # Fourth: Conv2DTranspose layer
    # 필터의 수는 64개, 커널 크기는 5, 스트라이드는 2, 패딩을 사용합니다.
    # 이미지를 확대하면서 64개의 특징 맵을 생성합니다.
    model.add(layers.Conv2DTranspose(64, kernel_size=5, strides=2, padding='same', use_bias=False))
    # 배치 정규화를 사용하여 데이터 분포를 정규화합니다.
    model.add(layers.BatchNormalization())
    # LeakyReLU 활성화 함수로 비선형성을 추가합니다.
    model.add(layers.LeakyReLU())

    # Fifth: Conv2DTranspose layer
    # 최종 출력층, 필터 수는 3 (RGB 채널), 커널 크기는 5, 스트라이드는 2, 패딩은 사용합니다.
    # 'tanh' 활성화 함수로 결과 값을 -1에서 1 사이로 정규화하여 이미지 데이터를 생성합니다.
    model.add(layers.Conv2DTranspose(3, kernel_size=(5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))

    return model  # 생성자 모델을 반환합니다.
```

```python
# 생성자 모델을 생성합니다.
generator = make_generator_model()

# 생성된 모델의 요약 정보를 출력합니다.
# 모델의 각 층의 이름, 출력 형태, 그리고 파라미터 수를 보여줍니다.
# 이를 통해 모델 구조를 확인하고, 층의 구성 및 파라미터 개수를 점검할 수 있습니다.
generator.summary()
```

```python
# 생성자 모델에 입력할 노이즈 벡터를 생성합니다.
# 크기가 [1, 100]인 텐서를 생성하며, 각 요소는 평균이 0이고 표준편차가 1인 정규분포에서 샘플링됩니다.
# 이는 GAN에서 무작위 노이즈 입력을 통해 가짜 이미지를 생성하기 위한 입력으로 사용됩니다.
noise = tf.random.normal([1, 100])
```

```python
# 생성자 모델을 사용하여 노이즈 벡터를 입력으로 받아 가짜 이미지를 생성합니다.
# training=False는 생성 과정에서 학습 모드가 아님을 명시하여 배치 정규화 등의 레이어가 추론 모드로 작동하도록 합니다.
generated_image = generator(noise, training=False)

# 생성된 이미지의 형태(차원)를 출력합니다.
# 예를 들어, (1, 32, 32, 3)이라면 32x32 크기의 RGB 이미지가 한 개 생성된 것을 의미합니다.
generated_image.shape
```

### TEP 4. 판별자 모델 구현하기

```python
def make_discriminator_model():
    # 판별자 모델을 만듭니다.

    # Start
    model = tf.keras.Sequential()  # 순차적으로 층을 쌓는 Keras Sequential 모델을 생성합니다.

    # First: Conv2D Layer
    # 필터의 크기는 5이며 64개 사용, stride는 2, 패딩은 'same'으로 사용합니다.
    # 입력 값의 크기는 32x32 크기의 RGB 이미지입니다.
    model.add(layers.Conv2D(64, kernel_size=5, strides=2, padding='same', input_shape=(32, 32, 3)))
    # LeakyReLU 활성화 함수 사용으로 음수 값에 대해 작은 기울기를 허용합니다.
    model.add(layers.LeakyReLU())
    # Dropout을 사용하여 노드의 30%를 무작위로 사용하지 않음으로써 과적합을 방지합니다.
    model.add(layers.Dropout(0.3))

    # Second: Conv2D Layer
    # 필터의 크기는 5이며 128개 사용, stride는 2, 패딩은 'same'으로 사용합니다.
    # 이 층에서는 특징 맵의 크기를 줄이면서 특징을 추출합니다.
    model.add(layers.Conv2D(128, kernel_size=5, strides=2, padding='same'))
    # LeakyReLU 활성화 함수 사용으로 비선형성을 추가합니다.
    model.add(layers.LeakyReLU())
    # Dropout을 사용하여 노드의 30%를 무작위로 사용하지 않음으로써 과적합을 방지합니다.
    model.add(layers.Dropout(0.3))

    # Third: Flatten Layer
    # 이미지를 1차원 벡터로 변환합니다. 이 과정은 Dense 층에 입력으로 사용하기 위해 필요합니다.
    model.add(layers.Flatten())

    # Fourth: Dense Layer
    # 출력층으로, 노드 수는 1개입니다. 이 노드는 입력 이미지가 진짜인지 가짜인지에 대한 점수를 출력합니다.
    model.add(layers.Dense(1))

    return model  # 판별자 모델을 반환합니다.
```

```python
# 판별자 모델을 생성합니다.
discriminator = make_discriminator_model()

# 생성된 판별자 모델의 요약 정보를 출력합니다.
# 모델의 각 층의 이름, 출력 형태, 그리고 파라미터 수를 보여줍니다.
# 이를 통해 모델 구조를 확인하고, 층의 구성 및 파라미터 개수를 점검할 수 있습니다.
discriminator.summary()
```

```python
# 판별자 모델을 사용하여 생성된 이미지가 진짜인지 가짜인지에 대한 결정을 내립니다.
# training=False는 판별자 모델이 학습 모드가 아님을 명시하여, 배치 정규화 등의 층이 추론 모드로 작동하도록 합니다.
decision = discriminator(generated_image, training=False)
```

### STEP 5. 손실함수와 최적화 함수 구현하기

```python
# 교차 엔트로피 손실 함수를 정의합니다.
# BinaryCrossentropy는 이진 분류 문제에 적합한 손실 함수입니다.
# from_logits=True는 출력이 확률이 아니라 로짓(실수 값)임을 나타내며, 이 경우 시그모이드 함수가 자동으로 적용됩니다.
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)
```

```python
def generator_loss(fake_output):
    # 생성자의 손실 함수를 정의합니다.
    # 생성자는 판별자가 생성된 이미지(fake_output)를 진짜(1)로 잘못 판단하게 하려고 합니다.
    # 이를 위해 모든 출력이 1이 되도록 하는 손실을 계산합니다.
    return cross_entropy(tf.ones_like(fake_output), fake_output)
```

```python
def discriminator_loss(real_output, fake_output):
    # 판별자의 손실 함수를 정의합니다.
    
    # 진짜 이미지에 대한 손실을 계산합니다.
    # real_output은 진짜 이미지에 대한 판별자의 예측 결과입니다.
    # tf.ones_like(real_output)은 진짜 이미지를 1로 분류하도록 유도합니다.
    # 따라서 교차 엔트로피를 통해 진짜 이미지가 진짜로 분류될 수 있도록 손실을 계산합니다.
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    
    # 가짜 이미지에 대한 손실을 계산합니다.
    # fake_output은 생성자가 만든 가짜 이미지에 대한 판별자의 예측 결과입니다.
    # tf.zeros_like(fake_output)은 가짜 이미지를 0으로 분류하도록 유도합니다.
    # 교차 엔트로피를 통해 가짜 이미지가 가짜로 분류될 수 있도록 손실을 계산합니다.
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    
    # 진짜 이미지와 가짜 이미지에 대한 손실을 합산하여 판별자의 총 손실을 계산합니다.
    total_loss = real_loss + fake_loss
    
    return total_loss  # 총 손실을 반환합니다.
```

```python
def discriminator_accuracy(real_output, fake_output):
    # 판별자의 정확도를 계산하는 함수입니다.
    
    # 진짜 이미지에 대한 정확도를 계산합니다.
    # real_output의 값이 0.5 이상인 경우를 진짜로 판별한 것으로 간주합니다.
    # tf.math.greater_equal을 사용하여 real_output >= 0.5인 경우를 True로 반환합니다.
    # tf.cast를 통해 이를 float32 형으로 변환하고, tf.reduce_mean으로 평균을 구해 정확도를 계산합니다.
    real_accuracy = tf.reduce_mean(tf.cast(tf.math.greater_equal(real_output, tf.constant([0.5])), tf.float32))
    
    # 가짜 이미지에 대한 정확도를 계산합니다.
    # fake_output의 값이 0.5 미만인 경우를 가짜로 판별한 것으로 간주합니다.
    # tf.math.less를 사용하여 fake_output < 0.5인 경우를 True로 반환합니다.
    # tf.cast를 통해 이를 float32 형으로 변환하고, tf.reduce_mean으로 평균을 구해 정확도를 계산합니다.
    fake_accuracy = tf.reduce_mean(tf.cast(tf.math.less(fake_output, tf.constant([0.5])), tf.float32))
    
    return real_accuracy, fake_accuracy  # 진짜 이미지와 가짜 이미지에 대한 정확도를 반환합니다.
```

```python
# 생성자 모델의 옵티마이저를 정의합니다.
# Adam 옵티마이저를 사용하며, 학습률(learning rate)은 1e-4로 설정합니다.
generator_optimizer = tf.keras.optimizers.Adam(1e-4)

# 판별자 모델의 옵티마이저를 정의합니다.
# Adam 옵티마이저를 사용하며, 학습률은 1e-4로 설정합니다.
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)
```

```python
def generate_and_save_images(model, epoch, it, sample_seeds):
    # 특정 에포크와 반복(iteration)에서 이미지 생성 및 저장하는 함수입니다.

    # 생성자 모델을 사용하여 샘플 시드로부터 이미지를 생성합니다.
    # training=False는 학습 모드가 아님을 명시하여, 배치 정규화 등의 층이 추론 모드로 작동하도록 합니다.
    predictions = model(sample_seeds, training=False)

    # 이미지들을 4x4 격자 형태로 출력합니다.
    fig = plt.figure(figsize=(4, 4))
    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i + 1)  # 4x4 그리드에서 i+1번째 서브플롯을 설정합니다.
        # 생성된 이미지의 픽셀 값을 0~255 사이로 변환하여 시각화합니다.
        # predictions의 출력 값이 tanh 활성화 함수로 -1에서 1 사이에 있으므로,
        # 이를 원래의 0~255 범위로 되돌리기 위해 (127.5를 곱하고 127.5를 더합니다.)
        plt.imshow(np.array(predictions[i] * 127.5 + 127.5).astype('uint8'))
        plt.axis('off')  # 각 이미지의 축을 표시하지 않습니다.

    # 생성된 이미지를 파일로 저장합니다.
    # 파일 이름에 에포크와 반복 번호를 포함하여 저장합니다.
    plt.savefig('./generated_samples/sample_epoch_{:04d}_iter_{:03d}.png'.format(epoch, it))

    # 생성된 이미지를 화면에 표시합니다.
    plt.show()
```

### STEP 7. 학습 과정 진행하기

```python
def train(dataset, epochs, save_every):
    # 모델을 학습하는 함수입니다.
    # dataset: 학습에 사용할 데이터셋입니다.
    # epochs: 총 학습 에포크 수입니다.
    # save_every: 모델을 저장할 에포크 간격입니다.
    
    start = time.time()  # 학습 시작 시간을 기록합니다.
    history = {'gen_loss': [], 'disc_loss': [], 'real_accuracy': [], 'fake_accuracy': []}  # 손실과 정확도 이력을 저장할 딕셔너리입니다.

    for epoch in range(epochs):
        # 에포크 시작 시간을 기록합니다.
        epoch_start = time.time()

        # 데이터셋을 반복하며 배치 단위로 학습을 진행합니다.
        for it, image_batch in enumerate(dataset):
            # 한 학습 스텝을 진행하고 생성자와 판별자의 손실 및 판별자의 정확도를 반환받습니다.
            gen_loss, disc_loss, real_accuracy, fake_accuracy = train_step(image_batch)
            
            # 학습 이력을 기록합니다.
            history['gen_loss'].append(gen_loss)
            history['disc_loss'].append(disc_loss)
            history['real_accuracy'].append(real_accuracy)
            history['fake_accuracy'].append(fake_accuracy)

            # 50번째 반복마다 중간 결과를 시각화하고 현재 학습 상태를 출력합니다.
            if it % 50 == 0:
                display.clear_output(wait=True)  # 출력 화면을 지우고 업데이트된 결과를 표시합니다.
                generate_and_save_images(generator, epoch + 1, it + 1, seed)  # 생성된 이미지를 저장하고 화면에 표시합니다.
                print('Epoch {} | iter {}'.format(epoch + 1, it + 1))  # 현재 에포크와 반복 회차를 출력합니다.
                print('Time for epoch {} : {} sec'.format(epoch + 1, int(time.time() - epoch_start)))  # 현재 에포크의 경과 시간을 출력합니다.

        # 지정된 에포크 간격마다 체크포인트를 저장합니다.
        if (epoch + 1) % save_every == 0:
            checkpoint.save(file_prefix=checkpoint_prefix)  # 모델의 상태를 체크포인트로 저장합니다.

        # 에포크가 끝날 때마다 최종 상태를 시각화합니다.
        display.clear_output(wait=True)
        generate_and_save_images(generator, epochs, it, seed)  # 에포크가 종료된 후 최종 이미지를 생성하고 저장합니다.
        print('Time for training : {} sec'.format(int(time.time() - start)))  # 전체 학습 시간 출력합니다.

        # 학습 이력 그래프를 그립니다.
        draw_train_history(history, epoch)

```

```python
save_every = 5  # 모델 체크포인트를 저장할 에포크 간격을 설정합니다. 여기서는 5 에포크마다 모델 상태를 저장합니다.

EPOCHS = 30  # 총 학습 에포크 수를 30으로 설정합니다.
```

```python
def make_generator_model4():
    # 생성자 모델을 만드는 함수입니다.

    model = tf.keras.Sequential()  # Sequential 모델을 사용하여 층을 순차적으로 쌓습니다.

    # 첫 번째 층: Dense Layer
    # 입력은 100차원의 벡터이며, 이를 8x8x128 차원의 텐서로 변환합니다.
    # ReLU 활성화 함수가 사용됩니다.
    model.add(layers.Dense(8 * 8 * 128, activation='relu', input_shape=(100,)))
    # 배치 정규화를 사용하여 학습의 안정성을 높입니다. 모멘텀(momentum)은 0.9로 설정됩니다.
    model.add(layers.BatchNormalization(momentum=0.9))
    # LeakyReLU 활성화 함수를 사용하여 음수 입력에 대해 작은 기울기를 허용합니다. 알파 값은 0.1입니다.
    model.add(layers.LeakyReLU(alpha=0.1))
    
    # Reshape Layer
    # Dense 층의 출력을 8x8 크기의 128개의 채널로 재구성합니다.
    model.add(layers.Reshape((8, 8, 128)))

    # 첫 번째 Conv2D Layer
    # 커널 크기는 4x4, stride는 1, 'same' 패딩을 사용하여 출력 크기를 유지합니다.
    model.add(layers.Conv2D(128, kernel_size=4, strides=1, padding='same'))
    # 배치 정규화를 사용합니다.
    model.add(layers.BatchNormalization(momentum=0.9))
    # LeakyReLU 활성화 함수 사용
    model.add(layers.LeakyReLU(alpha=0.1))

    # 첫 번째 Conv2DTranspose Layer
    # 커널 크기는 4x4, stride는 2, 'same' 패딩을 사용하여 이미지 크기를 두 배로 확대합니다.
    model.add(layers.Conv2DTranspose(128, kernel_size=4, strides=2, padding='same'))
    # 배치 정규화를 사용합니다.
    model.add(layers.BatchNormalization(momentum=0.9))
    # LeakyReLU 활성화 함수 사용
    model.add(layers.LeakyReLU(alpha=0.1))

    # 두 번째 Conv2D Layer
    # 커널 크기는 5x5, stride는 1, 'same' 패딩을 사용하여 출력 크기를 유지합니다.
    model.add(layers.Conv2D(128, kernel_size=5, strides=1, padding='same'))
    # 배치 정규화를 사용합니다.
    model.add(layers.BatchNormalization(momentum=0.9))
    # LeakyReLU 활성화 함수 사용
    model.add(layers.LeakyReLU(alpha=0.1))
    
    # 두 번째 Conv2DTranspose Layer
    # 커널 크기는 4x4, stride는 2, 'same' 패딩을 사용하여 이미지 크기를 두 배로 확대합니다.
    model.add(layers.Conv2DTranspose(128, kernel_size=4, strides=2, padding='same'))
    # 배치 정규화를 사용합니다.
    model.add(layers.BatchNormalization(momentum=0.9))
    # LeakyReLU 활성화 함수 사용
    model.add(layers.LeakyReLU(alpha=0.1))

    # 세 번째 Conv2D Layer
    # 커널 크기는 5x5, stride는 1, 'same' 패딩을 사용하여 출력 크기를 유지합니다.
    model.add(layers.Conv2D(128, kernel_size=5, strides=1, padding='same'))
    # 배치 정규화를 사용합니다.
    model.add(layers.BatchNormalization(momentum=0.9))
    # LeakyReLU 활성화 함수 사용
    model.add(layers.LeakyReLU(alpha=0.1))

    # 출력 Conv2D Layer
    # 마지막 출력 층으로, 필터 수는 3개 (RGB 채널), 커널 크기는 5x5, stride는 1, 'same' 패딩을 사용합니다.
    # 'tanh' 활성화 함수를 사용하여 출력 픽셀 값을 -1에서 1 사이로 정규화합니다.
    model.add(layers.Conv2D(3, kernel_size=5, strides=1, padding='same', activation='tanh'))
    
    return model  # 생성된 생성자 모델을 반환합니다.
```

```python
# 생성자 모델4에 입력할 노이즈 벡터를 생성합니다.
# 크기는 (1, 100)이며, 각 요소는 평균이 0이고 표준편차가 1인 정규분포에서 샘플링됩니다.
# 1은 배치 크기를 의미하며, 100은 생성자 모델의 입력 차원입니다.
noise = tf.random.normal((1, 100))

# 생성된 노이즈를 생성자 모델에 입력하여 가짜 이미지를 생성합니다.
# training=False는 학습 모드가 아님을 명시하여, 배치 정규화 등의 층이 추론 모드로 작동하도록 합니다.
generated_image = generator4(noise, training=False)

# 생성된 이미지의 형태(차원)를 출력합니다.
# 보통 생성자는 32x32 크기의 RGB 이미지를 출력하며, 출력 형태는 (1, 32, 32, 3)와 같습니다.
generated_image.shape
```

```python
def make_discriminator_model4():
    # 판별자 모델을 만드는 함수입니다.

    model = tf.keras.Sequential()  # Sequential 모델을 사용하여 층을 순차적으로 쌓습니다.

    # 첫 번째 Conv2D Layer
    # 입력은 32x32 크기의 RGB 이미지입니다. 필터 수는 128개이며, 커널 크기는 3x3, 스트라이드는 1로 설정됩니다.
    # 패딩은 'same'으로 설정되어 출력 크기를 유지합니다.
    model.add(layers.Conv2D(128, kernel_size=3, strides=1, input_shape=(32, 32, 3), padding='same'))
    # 배치 정규화를 사용하여 학습의 안정성을 높입니다. 모멘텀(momentum)은 0.9로 설정됩니다.
    model.add(layers.BatchNormalization(momentum=0.9))
    # LeakyReLU 활성화 함수를 사용하여 음수 값에 대한 작은 기울기를 허용합니다. 알파 값은 0.1입니다.
    model.add(layers.LeakyReLU(alpha=0.1))
    # 드롭아웃을 사용하여 노드의 25%를 무작위로 사용하지 않음으로써 과적합을 방지합니다.
    model.add(layers.Dropout(0.25))

    # 두 번째 Conv2D Layer
    # 필터 수는 128개이며, 커널 크기는 4x4, 스트라이드는 2로 설정하여 이미지 크기를 줄입니다.
    model.add(layers.Conv2D(128, kernel_size=4, strides=2, padding='same'))
    model.add(layers.BatchNormalization(momentum=0.9))
    # LeakyReLU 활성화 함수 사용, 알파 값은 0.2입니다.
    model.add(layers.LeakyReLU(alpha=0.2))
    model.add(layers.Dropout(0.25))

    # 세 번째 Conv2D Layer
    # 필터 수는 128개이며, 커널 크기는 4x4, 스트라이드는 2로 설정하여 이미지 크기를 줄입니다.
    model.add(layers.Conv2D(128, kernel_size=4, strides=2, padding='same'))
    model.add(layers.BatchNormalization(momentum=0.9))
    model.add(layers.LeakyReLU(alpha=0.2))
    model.add(layers.Dropout(0.25))

    # 네 번째 Conv2D Layer
    # 필터 수는 128개이며, 커널 크기는 4x4, 스트라이드는 2로 설정하여 이미지 크기를 줄입니다.
    model.add(layers.Conv2D(128, kernel_size=4, strides=2, padding='same'))
    model.add(layers.BatchNormalization(momentum=0.9))
    model.add(layers.LeakyReLU(alpha=0.2))
    model.add(layers.Dropout(0.25))

    # Flatten Layer
    # 2D 형태의 데이터를 1D 벡터로 변환하여 Fully Connected Layer에 입력할 수 있도록 합니다.
    model.add(layers.Flatten())

    # 첫 번째 Dense Layer
    # 512개의 노드를 가진 완전 연결층입니다. ReLU 활성화 함수를 사용하여 비선형성을 추가합니다.
    model.add(layers.Dense(512, activation='relu'))

    # 출력 Dense Layer
    # 1개의 노드를 가진 출력층으로, 시그모이드 활성화 함수를 사용하여 출력값을 0과 1 사이의 확률로 만듭니다.
    model.add(layers.Dense(1, activation='sigmoid'))

    return model  # 생성된 판별자 모델을 반환합니다.
```
